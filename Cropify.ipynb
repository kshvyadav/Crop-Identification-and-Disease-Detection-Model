```python
import tkinter as tk
from tkinter import *
from tkinter import filedialog
from PIL import Image, ImageTk
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as pl
```

Crop Identification Model

```python
resnetcropidentificationmodel = tf.keras.models.load_model('Crop Identification Models/Resnet50Model.keras')
```

Disease Detection Models

```python
corndiseasedetectionmodel = tf.keras.models.load_model('Disease Detection Models/Corn Disease Detection Model/CornDiseaseDetectionModel.keras')
grapediseasedetectionmodel = tf.keras.models.load_model('Disease Detection Models/Grape Disease Detection Model/GrapeDiseaseDetectionModel.keras')
potatodiseasedetectionmodel = tf.keras.models.load_model('Disease Detection Models/Potato Disease Detection Model/PotatoDiseaseDetectionModel.keras')
tomatodiseasedetectionmodel = tf.keras.models.load_model('Disease Detection Models/Tomato Disease Detection Model/TomatoDiseaseDetectionModel.keras')
```

Creating python based application using tkinter lib to test our model prediction 

```python

root = tk.Tk()
root.title("Cropify")
root.geometry("678x336")  
root.resizable(False, False)  

background_image_path = "CBG.jpg"  
bg_image = Image.open(background_image_path)
bg_photo = ImageTk.PhotoImage(bg_image)

background_label = tk.Label(root, image=bg_photo)
background_label.place(x=0, y=0, relwidth=1, relheight=1)  

btn_image_path = "btnimg.jpg"  
btn_image = Image.open(btn_image_path)
btn_photo = ImageTk.PhotoImage(btn_image)



def open():
    global filename
    global diseasedetectionmodel_output_class
    global cropidentificatonmodel_output_class
    filename = filedialog.askopenfilename(filetypes=[("Image Files", "*.png;*.jpg;*.jpeg;*.bmp;*.gif")])
    
    image_path = filename
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title('Test Image')
    plt.xticks([])
    plt.yticks([])
    plt.show()

    img_resized = cv2.resize(img,(224,224))    
    img = np.expand_dims(img_resized,axis=0)
    print(img.shape)

    # global cropidentificatonmodel_output_class
    validation_set = keras.utils.image_dataset_from_directory(
    directory = 'Crop Identification Datasets/valid')

    class_name = validation_set.class_names
    print(class_name)

    resnetcropidentificationmodel_prediction=resnetcropidentificationmodel.predict(img)
    
    cropidentificatonmodel_output_class = class_name[np.argmax(resnetcropidentificationmodel_prediction)]
    confidence = round(100 * (np.max(resnetcropidentificationmodel_prediction)))
    print("Crop identification = ",cropidentificatonmodel_output_class)
    print(confidence)

        
    

    if cropidentificatonmodel_output_class == 'Tomato':
        validation_set = keras.utils.image_dataset_from_directory(
        directory = 'Disease Detection Datasets/Tomato/valid')

        class_name = validation_set.class_names
        print(class_name)

        tomatodiseasedetectionmodel_prediction=tomatodiseasedetectionmodel.predict(img)
    
        diseasedetectionmodel_output_class = class_name[np.argmax(tomatodiseasedetectionmodel_prediction)]
        print("Disease Detected = ",diseasedetectionmodel_output_class)
        confidence = round(100 * (np.max(tomatodiseasedetectionmodel_prediction)))
        print(confidence)
        output()
    

    elif cropidentificatonmodel_output_class == 'Corn':
        validation_set = keras.utils.image_dataset_from_directory(
        directory = 'Disease Detection Datasets/Corn/valid')

        class_name = validation_set.class_names
        print(class_name)

        corndiseasedetectionmodel_prediction=corndiseasedetectionmodel.predict(img)
    
        diseasedetectionmodel_output_class = class_name[np.argmax(corndiseasedetectionmodel_prediction)]
        print("Disease Detected = ",diseasedetectionmodel_output_class)
        confidence = round(100 * (np.max(corndiseasedetectionmodel_prediction)))
        print(confidence)
        output()


    elif cropidentificatonmodel_output_class == 'Grape':
        validation_set = keras.utils.image_dataset_from_directory(
        directory = 'Disease Detection Datasets/Grape/valid')

        class_name = validation_set.class_names
        print(class_name)

        grapediseasedetectionmodel_prediction=grapediseasedetectionmodel.predict(img)
    
        diseasedetectionmodel_output_class = class_name[np.argmax(grapediseasedetectionmodel_prediction)]
        print("Disease Detected = ",diseasedetectionmodel_output_class)
        confidence = round(100 * (np.max(grapediseasedetectionmodel_prediction)))
        print(confidence)
        output()
    

    elif cropidentificatonmodel_output_class == 'Potato':
        validation_set = keras.utils.image_dataset_from_directory(
        directory = 'Disease Detection Datasets/Potato/valid')

        class_name = validation_set.class_names
        print(class_name)

        potatodiseasedetectionmodel_prediction=potatodiseasedetectionmodel.predict(img)
    
        diseasedetectionmodel_output_class = class_name[np.argmax(potatodiseasedetectionmodel_prediction)]
        print("Disease Detected = ",diseasedetectionmodel_output_class)
        confidence = round(100 * (np.max(potatodiseasedetectionmodel_prediction)))
        print(confidence)
        output()

def output():

    popupwindow = Toplevel(root)
    popupwindow.title("Prediction")
    popupwindow.geometry("300x100")
    crop_identification = Label(popupwindow,text=f"Crop identification = {cropidentificatonmodel_output_class}")
    disease_detection = Label(popupwindow,text=f"Disease Detected = {diseasedetectionmodel_output_class}")
    btn = Button (popupwindow , text="Exit", command=popupwindow.destroy)
    crop_identification.pack()
    disease_detection.pack()
    btn.pack()
    popupwindow.mainloop()
    




select_button = tk.Button(root,image=btn_photo,bg="#9CCBC5",font="bold 20",borderwidth=0,command=open)
select_button.place(x=500, y=270)


root.mainloop()
```
